```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
link = "/delhi_aqi_day_wise.csv"
data = pd.read_csv(link)
```

```python
print(data.describe())
print(data.iloc[5])
print(data["PM2.5"])
```

```python
print(data.isnull().sum())
data.dropna(inplace=True)
print(data.isnull().sum())
print(data.describe())
```

```python
# Convert date strings to datetime objects
data['Date'] = pd.to_datetime(data['Date'], format="%d-%m-%Y")

# Create a new column 'Season' based on the 'Date' column
data['Season'] = data['Date'].dt.month.map({1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Autumn', 10: 'Autumn', 11: 'Autumn', 12: 'Winter'})
```

```python
# Plotting the line graph for PM2.5 values
plt.plot(data['Date'], data["PM2.5"])

# Adding labels and title
plt.xlabel('DATE')
plt.ylabel('PM2.5 VALUES (units = ug/m3)')
plt.title('Amount of PM2.5 in atomosphere from 2015-2020')

# Displaying the graph
plt.show()
```
